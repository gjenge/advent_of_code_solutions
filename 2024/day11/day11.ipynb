```python
def transform(stone: int):
    if stone == 0:
        return [1]
    elif len(str(stone)) % 2 == 0:
        stone_string = str(stone)
        left = stone_string[0 : len(stone_string)//2]
        right = stone_string[len(stone_string)//2:]
        return [int(left), int(right)]
    else:
        return [stone*2024]

def blink(stones: dict, saved_result: dict): 
    blinked_stones = stones.copy()
    for stone in {s for s, count in stones.items() if count > 0}: 
        if stone not in saved_result.keys():
            saved_result[stone] = transform(stone)
        for s in saved_result[stone]:
            if s not in blinked_stones.keys():
                blinked_stones[s] = 0   
            blinked_stones[s] += stones[stone]
        blinked_stones[stone] -= stones[stone]
    
    return blinked_stones
  
with open("input") as f:
    line = f.readline()
    input = [int(n) for n in line.strip().split(" ")]
    
saved_result = dict()
stones = dict()
for stone in input:
    stones[stone] = 1
for _ in range(75):
    stones = blink(stones, saved_result)

num_stones = 0
for num in stones.values():
    if num > 0:
        num_stones += num

num_stones
```
