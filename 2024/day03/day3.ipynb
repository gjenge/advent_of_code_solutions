```python
#PART 2
result = 0
state = "start"
do = True
num1_string = ""
num2_string = ""

def restart(character: str):
    if character == "m":
        return "m"
    elif character == "d":
        return "d"
    else:
        return "start"

with open("input") as input:
    character = input.read(1)
    while character:
        if state == "start":
            if character == "m":
                state = "m"
            elif character == "d":
                state = "d"
                
        elif state == "m":
            if character == "u":
                state = "u"
            else:
                state = restart(character)
                
        elif state == "u":
            if character == "l":
                state = "l"
            else:
                state = restart(character)
                
        elif state == "l":
            if character == "(":
                state = "("
            else:
                state = restart(character)
                
        elif state == "(":
            if character.isdigit():
                state = "num1_1"
                num1_string += character
            else:
                state = restart(character)
                num1_string = ""
                num2_string = ""
                
        elif state == "num1_1":
            if character.isdigit():
                state = "num1_2"
                num1_string += character
            elif character == ",":
                state = ","
            else:
                state = restart(character)
                num1_string = ""
                num2_string = ""
                
        elif state == "num1_2":
            if character.isdigit():
                state = "num1_3"
                num1_string += character
            elif character == ",":
                state = ","
            else:
                state = restart(character)
                num1_string = ""
                num2_string = ""
                
        elif state == "num1_3":
            if character == ",":
                state = ","
            else:
                state = restart(character)
                num1_string = ""
                num2_string = ""
                
        elif state == ",":
            if character.isdigit():
                state = "num2_1"
                num2_string += character
            else:
                state = restart(character)
                num1_string = ""
                num2_string = ""
                
        elif state == "num2_1":
            if character.isdigit():
                state = "num2_2"
                num2_string += character
            elif character == ")":
                state = ")"
            else:
                state = restart(character)
                num1_string = ""
                num2_string = ""
                
        elif state == "num2_2":
            if character.isdigit():
                state = "num2_3"
                num2_string += character
            elif character == ")":
                state = ")"
            else:
                state = restart(character)
                num1_string = ""
                num2_string = ""
                
        elif state == "num2_3":
            if character == ")":
                state = ")"
            else:
                state = restart(character)
                num1_string = ""
                num2_string = ""
                
        elif state == ")":
            num1 = int(num1_string)
            num2 = int(num2_string)
            if do:
                result += num1 * num2
            num1_string = ""
            num2_string = ""
            state = restart(character)
                
                
        elif state == "d":
            if character == "o":
                state = "o"
            else:
                state = restart(character)
                
        elif state == "o":
            if character == "(":
                state = "do("
            elif character == "n":
                state = "n"
            else:
                state = restart(character)
                
        elif state == "do(":
            if character == ")":
                state = "do)"
            else:
                state = restart(character)
                
        elif state == "do)":
            do = True
            state = restart(character)
            
        elif state == "n":
            if character == "'":
                state = "'"
            else:
                state = restart(character)
                
        elif state == "'":
            if character == "t":
                state = "t"
            else:
                state = restart(character)
                
        elif state == "t":
            if character == "(":
                state = "dont("
            else:
                state = restart(character)
                
        elif state == "dont(":
            if character == ")":
                state = "dont)"
            else:
                state = restart(character)
                
        elif state == "dont)":
            do = False
            state = restart(character)
            
        character = input.read(1)

result
            
        
```
