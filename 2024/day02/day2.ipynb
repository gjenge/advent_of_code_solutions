```python
#PART 1
safe_reports = 0
with open("input") as input:
    for line in input:
        report = [int(n) for n in line.split(" ")]
        increasing = False
        if report[0] < report[1]: #increasing
            increasing = True
        
        last = report[0]
        safe = True
        for i in range(1, len(report)):
            curr = report[i]
            distance = abs(last - curr)
            
            if (distance < 1 or distance > 3):
                safe = False
            if (increasing and last >= curr):
                safe = False
            if (not increasing and last <= curr):
                safe = False
            last = curr
        if safe:
            safe_reports += 1
safe_reports
                
        
        
```

```python
#PART 2
def check_report(report:list, levelAlreadyRemoved:bool):
    if len(report) == 1:
        return True
    increasing = False
    if report[0] < report[1]: #increasing
        increasing = True
        
    last = report[0]
    safe = True
    i = 1
    
    while i < len(report):
        curr = report[i]
        safe = check_elements(last, curr, increasing)
        
        #Try removal
        if not safe and not levelAlreadyRemoved and i < len(report)-1:
            return check_report(report[:i] + report[i+1:], levelAlreadyRemoved=True) or check_report(report[:i-1] + report[i:], levelAlreadyRemoved=True)
        
        if not safe:
            return False
            
        last = curr
        i+=1
        
    return safe

def check_elements(last:int, curr:int, increasing:bool):
    distance = abs(last - curr)
        
    if (distance < 1 or distance > 3):
        return False
    if (increasing and last >= curr):
        return False
    if (not increasing and last <= curr):
        return False
    
    return True

#main
safe_reports = 0
with open("input") as input:
    for line in input:
        report = [int(n) for n in line.split(" ")]
        if check_report(report, levelAlreadyRemoved=False):
            safe_reports +=1
            continue
        #Try removal of first element
        elif check_report(report[1:], levelAlreadyRemoved=True):
            safe_reports +=1
            continue
        #Try removal of last element
        elif check_report(report[:-1], levelAlreadyRemoved=True):
            safe_reports +=1
            continue
safe_reports
                
        
        
```
