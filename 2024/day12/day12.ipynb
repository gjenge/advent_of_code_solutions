```python
#region parametro di output {{plants}, area, perimeter}
def explore(seen_plants: set, plant_coords: tuple, input: list, plant_region: dict, region: dict) -> tuple:
    plant_region[plant_coords] = region
    region["plants"].add(plant_coords)
    region["area"] += 1
    seen_plants.add(plant_coords)
    i, j = plant_coords
    for neighbor in {(i-1,j), (i+1,j), (i,j-1), (i,j+1)}:
        #se il vicino non è fuori range dell'input e non è una pianta diversa e non si trova in una posizione che ho già visto
        if (neighbor[0] in range(len(input)) and neighbor[1] in range(len(input[0]))) and (input[neighbor[0]][neighbor[1]] == input[plant_coords[0]][plant_coords[1]]) and neighbor not in seen_plants:
            explore(seen_plants, neighbor, input, plant_region, region)

def find_row_perimeters(i: int, input: list, plant_region: dict): 
    prec_plant = ''
    for j, curr_plant in enumerate(input[i]):
        if prec_plant != curr_plant:
            keep_counting_up = True
            keep_counting_down = True
        #neighbor_up = (i-1, j)
        if not (i-1 in range(len(input))) or (input[i][j] != input[i-1][j]):
            if keep_counting_up:
                plant_region[(i,j)]["perimeter"] += 1
                keep_counting_up = False
        else:
            keep_counting_up = True
        #neighbor_down = (i+1, j)        
        if not (i+1 in range(len(input))) or (input[i][j] != input[i+1][j]):
            if keep_counting_down:
                plant_region[(i,j)]["perimeter"] += 1
                keep_counting_down = False
        else:
            keep_counting_down = True
        prec_plant = curr_plant 
        
def find_column_perimeters(j: int, input: list, plant_region: dict): 
    prec_plant = ''
    for i in range(len(input)):
        curr_plant = input[i][j]
        if prec_plant != curr_plant:
            keep_counting_right = True
            keep_counting_left = True
        #neighbor_left = (i, j-1)
        if not (j-1 in range(len(input[0]))) or (input[i][j] != input[i][j-1]):
            if keep_counting_left:
                plant_region[(i,j)]["perimeter"] += 1
                keep_counting_left = False
        else:
            keep_counting_left = True
        #neighbor_right = (i, j+1)
        if not (j+1 in range(len(input[0]))) or (input[i][j] != input[i][j+1]):
            if keep_counting_right:
                plant_region[(i,j)]["perimeter"] += 1
                keep_counting_right = False
        else:
            keep_counting_right = True
        prec_plant = curr_plant                                
                    
with open("input") as f:
    input = [[c for c in line.strip()] for line in f]


plant_region = dict()
seen_plants = set()
found_regions = list()
for i, row in enumerate(input):
    for j in range(len(row)):
        if (i,j) not in seen_plants:
            new_region = dict()
            new_region["plants"] = set()
            new_region["area"] = 0
            new_region["perimeter"] = 0
            explore(seen_plants, (i, j), input, plant_region, new_region)
            found_regions.append(new_region)

for i in range(len(input)):
    find_row_perimeters(i, input, plant_region)

for j in range(len(input[0])):
    find_column_perimeters(j, input, plant_region)

output = 0
for region in found_regions:
    print(region["area"], '*', region["perimeter"])
    output += region["area"] * region["perimeter"]
    
output
```
