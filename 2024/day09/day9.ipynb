```python
def next_free_space(index: int, disk: list) -> int:
    for i in range(index, len(disk)):
        if disk[i] == -1:
            return i
    return len(disk) - 1   

def next_chunk(index: int, disk: list) -> int:
    for i in range(index, -1, -1):
        if disk[i] != -1:
            return i
    return 0 

input = list()
with open("input") as f:
    line = f.readline()
    for ch in line:
        input.append(int(ch))
        
disk = list()
curr_id = 0
for i, num in enumerate(input):
    if i%2 == 0:
        for _ in range(num):
            disk.append(curr_id)
        curr_id += 1
    else:
        for _ in range(num):
            disk.append(-1)
            
#i starts at beginning and stops at free space, j starts at the end and stops at file chunks
i = next_free_space(0, disk)
j = next_chunk(len(disk) - 1, disk)
while i < j:
    tmp = disk[i]
    disk[i] = disk[j]
    disk[j] = tmp
    
    i = next_free_space(i+1, disk)
    j = next_chunk(j-1, disk)
    
checksum = 0
for i in range(0, len(disk)):
    if disk[i] != -1:
        checksum += (i * disk[i])
    
checksum
```

```python
def move_file(chunk_size: int, chunk_position: int, position: int, disk: list):
    for i, chunk_index in enumerate(range(chunk_position, chunk_position + chunk_size)):
        disk[position + i] = disk[chunk_index]
        disk[chunk_index] = -1

input = list()
with open("input") as f:
    line = f.readline()
    for ch in line:
        input.append(int(ch))
        
disk = list()
chunks = list()
free_spaces = list()
position_disk = 0
curr_id = 0
for i, num in enumerate(input):
    if i%2 == 0:
        chunks.append([position_disk,num])
        for _ in range(num):
            disk.append(curr_id)
        curr_id += 1
    else:
        free_spaces.append([position_disk,num])
        for _ in range(num):
            disk.append(-1)
    position_disk += num  

for j in range(len(chunks) - 1, -1, -1):
    chunk_size = chunks[j][1]
    chunk_position = chunks[j][0]
    for i in range(len(free_spaces)):
        fs_size = free_spaces[i][1]
        fs_position = free_spaces[i][0]
        if chunk_size <= fs_size and chunk_position >= fs_position:
            move_file(chunk_size, chunk_position, fs_position, disk)
            free_spaces[i][0] += chunk_size
            free_spaces[i][1] -= chunk_size
            free_spaces.append([chunk_position, chunk_size]) 
            break
        
checksum = 0
for i in range(len(disk)):
    if disk[i] != -1:
        checksum += (i * disk[i])

checksum
            
```
