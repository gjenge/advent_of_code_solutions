```python
def add_all(nums):
    ret = 0
    for n in nums:
        ret += n
    return ret
    
def mul_all(nums):
    ret = 1
    for n in nums:
        ret *= n
    return ret

output = 0
with open("input") as f:
    grid = list()
    for line in f:
        l = list(line)
        if l[len(l)-1] == '\n':
            l.pop()
        grid.append(l)

nums = []
operator = ''
for j in range(len(grid[0])-1, -1, -1):
    digits = []
    if operator == '+':
        output += add_all(nums)
        operator = ''
        nums.clear()
    if operator == '*':
        output += mul_all(nums)
        operator = ''
        nums.clear()
    for i in range(0, len(grid)):
        if  grid[i][j] == "+" or grid[i][j] == "*":
            operator = grid[i][j]
        elif grid[i][j] != ' ':
            digits.append(grid[i][j])
    if digits:
        nums.append(int(''.join(digits)))

if operator == '+':
    output += add_all(nums)
if operator == '*':
    output += mul_all(nums)
    
output
```
