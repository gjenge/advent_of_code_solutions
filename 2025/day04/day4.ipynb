```python
def count_adj_rolls(i: int, j: int, grid: list[list[chr]]) -> int:
    max_len_i = len(grid)
    max_len_j = len(grid[0])
    count = 0
    #generation of positions
    positions_to_check = []
    for tmp_i in range(i-1, i+2):
        for tmp_j in range(j-1, j+2):
            if i != tmp_i or j != tmp_j:
                positions_to_check.append((tmp_i, tmp_j))
    
    for pos in positions_to_check:
        tmp_i, tmp_j = pos
        if -1 < tmp_i < max_len_i and -1 < tmp_j < max_len_j and grid[tmp_i][tmp_j] == '@':
            count += 1
    
    return count

output = 0
with open("input") as f:
    grid = [list(line.strip()) for line in f]
    rolls_to_remove = []
    while True:
        for i in range(len(grid)):
            for j in range(len(grid[i])):
                if grid[i][j] == '@' and count_adj_rolls(i, j, grid) < 4:
                    rolls_to_remove.append((i, j))
        if not rolls_to_remove:
            break
        output += len(rolls_to_remove)
        
        for roll in rolls_to_remove:
            grid[roll[0]][roll[1]] = '.'
        rolls_to_remove.clear()

output   
```
