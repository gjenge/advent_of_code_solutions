```python
output = 0
with open("input") as f:
    ranges = []
    for line in f:
        if '-' in line:
            splitted = line.strip().split('-')
            ranges.append((int(splitted[0]), int(splitted[1])))
        elif line.strip() != "":
            id = int(line.strip())
            for r in ranges:
                if r[0] <= id <= r[1]:
                    output += 1
                    break            
                
output
```

```python
def overlap(x, y):
    return x[0] <= y[1] and y[0] <= x[1] #overlap        

output = 0
with open("input") as f:
    ranges = []
    mem_fresh = set()
    for line in f:
        mem_fresh.clear()
        if '-' in line:
            splitted = line.strip().split('-')
            curr = ((int(splitted[0]), int(splitted[1])))
            for r in ranges.copy():
                if(overlap(curr, r)):
                    curr = (min(curr[0], r[0]), max((curr[1], r[1])))
                    ranges.remove(r)
            ranges.append(curr)
    
    for r in ranges:
        output += (r[1] - r[0]) + 1
          
output
```
