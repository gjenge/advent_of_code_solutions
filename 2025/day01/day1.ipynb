```python
DIAL_RANGE=100
dial=50
password = 0

def rotate(dial: int, n: int)->int:
    tmp = dial + n
    res = abs(dial + n) % DIAL_RANGE
    if tmp >= 0:
        return res
    elif res == 0:
        return 0
    else:
        return DIAL_RANGE - res
    
with open("input") as f:
    for line in f:
        dir = line[0]
        n = int(line.strip()[1::])
        
        if dir == "R":
            dial = rotate(dial, n)
        else:
            dial = rotate(dial, -n)
            
        if dial == 0:
            password += 1
            
password
```

```python
DIAL_RANGE=100 
dial=50
password = 0

def rotate(dial: int, n: int)->tuple:
    res = abs(dial + n) % DIAL_RANGE
    click_0 = 0
    if n >= 0:
        click_0 = (dial + n) // DIAL_RANGE
    else:
        if(-n >= dial):
            if dial != 0:
                click_0 = 1
            click_0 += (-n - dial) // DIAL_RANGE
            if click_0 == 0 and dial != 0:
                click_0 = 1
                
    if dial + n >= 0:
        return (res, click_0)
    elif res == 0:
        return (0, click_0)
    else:
        return (DIAL_RANGE - res, click_0)
    
with open("input") as f:
    for line in f:
        dir = line[0]
        n = int(line.strip()[1::])
        if dir == "R":
            rotation_result = rotate(dial, n)
            dial = rotation_result[0]
        else:
            rotation_result = rotate(dial, -n)
            dial = rotation_result[0]
        password += rotation_result[1]
            
password
```
