```python
def find_max(seq: list[int])->tuple[int,int]:
    max = 0
    i_max = -1
    for i, val in enumerate(seq):
        if val > max:
            max = val
            i_max = i
    
    return (i_max, max)
        
output = 0        
with open("input") as f:
    for line in f:
        bank = [int(n) for n in line.strip()]
        first_index, first_number = find_max(bank[:len(bank)-1])
        second_number = find_max(bank[first_index+1:])[1]
        output += int(str(first_number) + str(second_number))

output
        
```

```python
def find_max(seq: list[int])->tuple[int,int]:
    max = 0
    i_max = -1
    for i, val in enumerate(seq):
        if val > max:
            max = val
            i_max = i
    
    return (i_max, max)
        
output = 0        
with open("input") as f:
    for line in f:
        bank = [int(n) for n in line.strip()]
        nums = []
        slice_index = -1
        for i in reversed(range(0, 12)):
            tmp_index, number = find_max(bank[slice_index+1:len(bank)-i])
            slice_index += tmp_index + 1
            nums.append(number)
        output += int(''.join(str(n) for n in nums))

output
        
```
