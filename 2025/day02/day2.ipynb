```python
output = 0

def check_parts_all_equal(parts: list)-> bool:
    prec = parts[0]
    for i in range(1, len(parts)):
        curr = parts[i]
        if prec != curr:
            return False
    return True

def check_invaild(num_str: str)-> bool:
    for slice_index in range(len(num_str)//2, 0, -1):
        if len(num_str) % slice_index == 0:
            parts = [num_str[i:i+slice_index] for i in range(0, len(num_str), slice_index)]
            if check_parts_all_equal(parts):
                return True
    return False
            

with open("input") as f:
    for line in f:
        intervals = line.strip().split(",")
        for interval in intervals:
            bounds = interval.split("-")
            lower_bound = int(bounds[0])
            upper_bound = int(bounds[1])
            for i in range(lower_bound, upper_bound+1):
                num_str = str(i)
                if check_invaild(num_str):
                    output += i
                    
output
                        
                
```
